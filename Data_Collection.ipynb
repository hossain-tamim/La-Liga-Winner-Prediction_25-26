```python
# # Define the list of teams
EloTeams = ['Alaves', 'Bilbao', 'Barcelona', 'Celta', 'Espanyol', 'Depor', 'Las Palmas', 'Mallorca', 'Malaga', 'Numancia', 'Osasuna', 'Oviedo', 'Santander', 'Rayo Vallecano', 'Real Madrid', 'Sociedad', 'Valencia', 'Valladolid', 'Villarreal', 'Zaragoza', 'Betis', 'Sevilla', 'Tenerife', 'Atletico', 'Recreativo', 'Albacete', 'Murcia', 'Getafe', 'Levante', 'Cadiz', 'Tarragona', 'Almeria', 'Gijon', 'Xerez', 'Hercules', 'Granada', 'Elche', 'Cordoba', 'Eibar', 'Leganes', 'Girona', 'Huesca']

#--------------------

print(len(EloTeams))

all_elo_data = []

for team in EloTeams:
    try:
        elo = sd.ClubElo() # Initialize ClubElo inside the loop to fetch data for each team
        team_elo_history = elo.read_team_history(team)
        all_elo_data.append(team_elo_history)  # Append the data to the list
        print(f"Got ELO data for {team}")
    except Exception as e:
        print(f"Failed to get ELO data for {team}: {e}")

if all_elo_data:
    EloRating = pd.concat(all_elo_data)
    print(f"Total ELO records: {EloRating.shape[0]}")

    # Only perform these operations if EloRating is not empty
    EloRating = EloRating.reset_index()


    display(EloRating.head())
else:
    print("No ELO data was retrieved.")
```

```python
import numpy as np

# Convert date columns in combined_match_stats to datetime if not already
combined_match_stats['date'] = pd.to_datetime(combined_match_stats['date'])
EloRating['from'] = pd.to_datetime(EloRating['from'])
EloRating['to'] = pd.to_datetime(EloRating['to'])

# Define team name mapping (ensure it covers the teams in combined_match_stats if needed)
# Based on the previous work, using the existing team_name_mapping
# You might need to add more mappings if there are teams in combined_match_stats
# that are not in the current team_name_mapping but are in EloRating with different names.
team_name_mapping = {
    'Athletic Club': 'Bilbao',
    'Celta Vigo': 'Celta',
    'Leganes': 'Leganes',
    'Mallorca': 'Mallorca',
    'Valencia': 'Valencia',
    'Villarreal': 'Villarreal',
    'Alaves': 'Alaves',
    'Atletico Madrid': 'Atletico',
    'Espanyol': 'Espanyol',
    'Real Betis': 'Betis',
    'Granada': 'Granada',
    'Levante': 'Levante',
    'Getafe': 'Getafe',
    'Osasuna': 'Osasuna',
    'Real Madrid': 'Real Madrid', # Corrected mapping for Real Madrid
    'Barcelona': 'Barcelona',
    'Sevilla': 'Sevilla',
    'Real Sociedad': 'Sociedad',
    'Eibar': 'Eibar',
    'Real Valladolid': 'Valladolid',
    'Cadiz': 'Cadiz',
    'SD Huesca': 'Huesca',
    'Elche': 'Elche',
    'Rayo Vallecano': 'Rayo Vallecano', # Corrected mapping for Rayo Vallecano
    'Almeria': 'Almeria',
    'Girona': 'Girona',
    'Las Palmas': 'Las Palmas' # Corrected mapping for Las Palmas
}


def get_elo_rating(team_name, match_date, elo_df, team_mapping):
    """
    Find the ELO rating for a team on a specific date
    """
    # Map team name if needed
    mapped_team = team_mapping.get(team_name, team_name)

    # Filter ELO data for the specific team
    team_elo = elo_df[elo_df['team'] == mapped_team].copy()

    if team_elo.empty:
        return np.nan

    # Find the rating valid on the match date
    # The rating is valid if match_date >= from and match_date <= to
    valid_ratings = team_elo[
        (team_elo['from'] <= match_date) &
        (team_elo['to'] >= match_date)
    ]

    if not valid_ratings.empty:
        # If multiple valid ratings, take the one with the latest 'from' date
        return valid_ratings.loc[valid_ratings['from'].idxmax(), 'elo']

    # If no exact match, find the closest rating before the match date
    before_match = team_elo[team_elo['to'] < match_date]
    if not before_match.empty:
        return before_match.loc[before_match['to'].idxmax(), 'elo']

    return np.nan


# Apply the function to get ELO ratings for home and away teams
print("Merging ELO ratings with combined_match_stats...")

combined_match_stats['home_team_elo'] = combined_match_stats.apply(
    lambda row: get_elo_rating(row['home_team'], row['date'], EloRating, team_name_mapping),
    axis=1
)

combined_match_stats['away_team_elo'] = combined_match_stats.apply(
    lambda row: get_elo_rating(row['away_team'], row['date'], EloRating, team_name_mapping),
    axis=1
)

print("ELO merge completed for combined_match_stats!")

# Check for missing ELO ratings
missing_home_elo = combined_match_stats['home_team_elo'].isna().sum()
missing_away_elo = combined_match_stats['away_team_elo'].isna().sum()

print(f"Number of matches missing home team ELO: {missing_home_elo}")
print(f"Number of matches missing away team ELO: {missing_away_elo}")

if missing_home_elo > 0:
    print("\nSample of matches missing home ELO:")
    display(combined_match_stats[combined_match_stats['home_team_elo'].isna()][['home_team', 'date']].drop_duplicates().head())

if missing_away_elo > 0:
    print("\nSample of matches missing away ELO:")
    display(combined_match_stats[combined_match_stats['away_team_elo'].isna()][['away_team', 'date']].drop_duplicates().head())


# Display sample results
print("\nSample of combined_match_stats with merged ELO ratings:")
display(combined_match_stats[['date', 'home_team', 'away_team', 'home_team_elo', 'away_team_elo']].head())
```

```python
from imblearn.over_sampling import RandomOverSampler

print("Distribution of the 'result' column:")
print(Team_match_statsB['result'].value_counts())

X = Team_match_statsB.drop(columns=['result'],axis = 1)

y = Team_match_statsB['result']


print("Target (y) shape:", y.shape)
print("Target (X) shape:", X.shape)


ros = RandomOverSampler(random_state=42)

X_resampled, y_resampled = ros.fit_resample(X, y)

print("\nDistribution of the resampled 'result' column:")
print(y_resampled.value_counts())
```

#Fetaure Engineering

```python
Team_match_statsB['Xg_diff'] = Team_match_statsB['home_xg'] - Team_match_statsB['away_xg']
Team_match_statsB['elo_diff'] = Team_match_statsB['home_team_elo'] - Team_match_statsB['away_team_elo']
Team_match_statsB['Xpoint_diff'] = Team_match_statsB['home_expected_points'] - Team_match_statsB['away_expected_points']
Team_match_statsB["ppda_ratio"] = Team_match_statsB["home_ppda"] / (Team_match_statsB["away_ppda"] + 1e-6)
Team_match_statsB["deep_comp_ratio"] = Team_match_statsB["home_deep_completions"] / (Team_match_statsB["away_deep_completions"] + 1e-6)
Team_match_statsB.drop(['away_points'],axis=1,inplace=True)
```

```python

resampled_team_match_statsB['Xg_diff'] = resampled_team_match_statsB['home_xg'] - resampled_team_match_statsB['away_xg']
resampled_team_match_statsB['elo_diff'] = resampled_team_match_statsB['home_team_elo'] - resampled_team_match_statsB['away_team_elo']
resampled_team_match_statsB['Xpoint_diff'] = resampled_team_match_statsB['home_expected_points'] - resampled_team_match_statsB['away_expected_points']
resampled_team_match_statsB["ppda_ratio"] = resampled_team_match_statsB["home_ppda"] / (resampled_team_match_statsB["away_ppda"] + 1e-6)
resampled_team_match_statsB["deep_comp_ratio"] = resampled_team_match_statsB["home_deep_completions"] / (resampled_team_match_statsB["away_deep_completions"] + 1e-6)
resampled_team_match_statsB.drop(['away_points'],axis=1,inplace=True)
```

```python
Team_match_statsB.sort_values('date', inplace=True)
# Group by team to avoid mixing home/away stats
for team_id in Team_match_statsB['home_team_id'].unique():
    # Home games
    home_mask = Team_match_statsB['home_team_id'] == team_id
    Team_match_statsB.loc[home_mask, 'home_xg_avg_last3'] = Team_match_statsB.loc[home_mask, 'home_xg'].rolling(3, min_periods=1).mean()

    # Away games
    away_mask = Team_match_statsB['away_team_id'] == team_id
    Team_match_statsB.loc[away_mask, 'away_xg_avg_last3'] = Team_match_statsB.loc[away_mask, 'away_xg'].rolling(3, min_periods=1).mean()

# Example: Momentum feature (improving/worsening form)
Team_match_statsB['home_xg_trend'] = Team_match_statsB.groupby('home_team_id')['home_xg_avg_last3'].diff()  # Upward trend = positive
```

```python
resampled_team_match_statsB.sort_values('date', inplace=True)
# Group by team to avoid mixing home/away stats
for team_id in resampled_team_match_statsB['home_team_id'].unique():
    # Home games
    home_mask = resampled_team_match_statsB['home_team_id'] == team_id
    resampled_team_match_statsB.loc[home_mask, 'home_xg_avg_last3'] = resampled_team_match_statsB.loc[home_mask, 'home_xg'].rolling(3, min_periods=1).mean()

    # Away games
    away_mask = resampled_team_match_statsB['away_team_id'] == team_id
    resampled_team_match_statsB.loc[away_mask, 'away_xg_avg_last3'] = resampled_team_match_statsB.loc[away_mask, 'away_xg'].rolling(3, min_periods=1).mean()

# Example: Momentum feature (improving/worsening form)
resampled_team_match_statsB['home_xg_trend'] = resampled_team_match_statsB.groupby('home_team_id')['home_xg_avg_last3'].diff()  # Upward trend = positive
```

```python
resampled_team_match_statsB['home_xg_trend'].fillna(resampled_team_match_statsB['home_xg_trend'].mean(), inplace=True)
print(resampled_team_match_statsB.isnull().sum())
Team_match_statsB['home_xg_trend'].fillna(Team_match_statsB['home_xg_trend'].mean(), inplace=True)
print('\n\n\n',Team_match_statsB.isnull().sum())


```
